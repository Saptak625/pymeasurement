```python
import pandas as pd

df = pd.read_excel('_static\Combustion_Lab_1920_Student_Data.xlsx')

# Go to src
import sys
import os
sys.path.insert(0, os.path.join(os.path.dirname('test.ipynb'), '..', 'src'))
from pymeasurement import Measurement as M, SigFig
```

```python
def convert(sample, uncertainty=None, units='', analog=False, digital=False, constant=False, u=None, a=False, d=False, un='', decimals=None):
  if u is not None:
    uncertainty = u
  if a:
    analog = a
  if d:
    digital = d
  if un:
    units = un
  
  return M(SigFig(str(sample), decimals=-decimals if decimals is not None else None), uncertainty=str(uncertainty) if uncertainty is not None else None, precision=float('inf') if constant else None, units=units, analog=analog, digital=digital)

decimals = [3, 3, 3, 1, 1]
units = ['g', 'g', 'g', 'ºC', 'ºC']
for i in range(5):
  df.iloc[:, i + 3] = df.iloc[:, i + 3].apply(convert, d=True, un=units[i], decimals=decimals[i])


df
```

```python
heat_capacity = M.fromStr('4.18c J/g*ºC')
j_to_kj = M.fromStr('0.001c kJ/J')
```

```python
df['Q of H2O (kJ)'] = df['Mass of water (+/- 0.001 g)'] * heat_capacity * (df['Final temperature (+/- 0.1 ºC)'] - df['Initial temperature (+/- 0.1 ºC)']) * j_to_kj
```

```python
df['Mass of Alcohol (g)'] = df['Initial mass of spirit burner (+/- 0.001 g)'] - df['Final mass of spirit burner (+/- 0.001 g)']
molar_masses = {'Ethanol': M.fromStr('46.08c g/mol'), 'Propan-1-ol' :M.fromStr('60.11c g/mol'), 'Butan-1-ol': M.fromStr('74.14c g/mol'), 'Pentan-1-ol': M.fromStr('86.17c g/mol')}
df['Molar mass of Alcohol (g/mol)'] = df.apply(lambda x: x['Mass of Alcohol (g)'] / molar_masses[x['Alcohol tested']], axis=1)
```

```python
df['Molar Enthalpy of Combustion (kJ/mol)'] = - df['Q of H2O (kJ)'] / df['Molar mass of Alcohol (g/mol)']

df
```

```python
results_df = pd.DataFrame()
results_df['Alcohol'] = df['Alcohol tested'].unique()
results_df['Average Molar Enthalpy of Combustion (kJ/mol)'] = M.average(df.groupby('Alcohol tested')['Molar Enthalpy of Combustion (kJ/mol)'])
```
